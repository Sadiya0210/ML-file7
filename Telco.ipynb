```python
# cleaning the dataset, converting
# totalcharges to numeric

df['totalcharges'] = pd.to_numeric(df['totalcharges'], errors='coerce')
```

```python
#droping the missing rows
df = df.dropna()
```

```python
df.shape
```

```python
# encoding the target variable
df['churn'] = df['churn'].map({'Yes': 1, 'No': 0})
```

```python
# feature target separation
# feature
x = df.drop(['churn','customerid'],axis=1)
#traget
y = df['churn']

#here 1 removed the unneccessary column customer
# id too
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# scaling in preprocessing is not needs for decision tree model

```

```python
cat_cols = x.select_dtypes(include='object').columns
num_cols = x.select_dtypes(exclude='object').columns

```

```python
# train test split
x_train,x_test,y_train,y_test=train_test_split(
    x,y,test_size=0.2,random_state=42,stratify=y
)

```

```python
# building sklearn encoder
preprocessor = ColumnTransformer(
    transformers=[
        ('cat',OneHotEncoder(drop='first',
          handle_unknown='ignore'),cat_cols),
        ('num','passthrough',num_cols)
    ]
)
```

```python
# now we will create a pipeline end to end for encoder

# on object, no leakage
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier
```

```python
 # pipeline = preprocessing + model together
# this avoids the means up in order

dt_pipeline = Pipeline(steps=[
        ('preprocessor',preprocessor),
        ('classifier',DecisionTreeClassifier(random_state=42))
    ])
```

```python
# Train the decision tree

dt_pipeline.fit(x_train,y_train)
```

```python
# Evaluating the model
from sklearn.metrics import accuracy_score,classification_report
y_pred_dt = dt_pipeline.predict(x_test)
#above one for test
y_pred_train = dt_pipeline.predict(x_train)
#above one for train
print(accuracy_score(y_train,y_pred_train))

print(accuracy_score(y_test,y_pred_dt))

#if train acc is hight , test acc is low_>overfitting
#if train acc is low like[50],
# test acc is low - underfitting
# if train and test acc is good enough
# and same - properfitting (if diffeerence is 2 then also)
```

```python
# classification report
print(classification_report(y_test,y_pred_dt))
```

```python
# overfitting intuition
for depth in[2,4,6,8,10,None]:
  dt_pipeline=Pipeline(steps=[
    ('preprocessor',preprocessor),
    ('model',DecisionTreeClassifier(max_depth=depth,random_state=42))
  ])
  dt_pipeline.fit(x_train,y_train)
  train_acc=dt_pipeline.score(x_train,y_train)
  test_acc=dt_pipeline.score(x_test,y_test)
  print(f'depth:{depth},train={train_acc:.2f},Test={test_acc:.2f}')
```

```python
# feature importance
dt_model=dt_pipeline.named_steps['model']
encoder=dt_pipeline.named_steps['preprocessor']

features_names=encoder.get_feature_names_out()

importances=pd.Series(
    dt_model.feature_importances_,
    index=features_names
).sort_values(ascending=False)

importances.head(10)
```

```python
 # feature importance

 # a pipeline that wraps everything together
 # to inspect internals
 # dt_model - the trained decision tree
 # encoder - the fitted columntransfermer
 # higher value more impoetant
 # values sum to 1.0
 # trees calculate this automatically during training
```

```python
# random forest
from sklearn.ensemble import RandomForestClassifier

rf_pipeline=Pipeline([
    ('preprocessor',preprocessor),
    ('model',RandomForestClassifier(
        n_estimators=200,
        random_state=42,
        max_depth=8,
        n_jobs=-1
    ))
])

rf_pipeline.fit(x_train,y_train)

```
